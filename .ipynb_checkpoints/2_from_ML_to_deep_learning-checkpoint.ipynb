# Logistic Classifier

### Softmax
- turn score to propability

สำหรับ algorithm classification ที่คำตอบจะเป็นค่าใน set คำตอบ(เช่น A B C)เท่านั้น softmax จะช่วยให้มีค่า probability แนบออกมาด้วย

เมื่อต้องการแปลงค่า (logit) score $y_i$ จากเซต $Y$ ให้อยู่ในรูป probability จะได้

$$S(y_i) = \frac{e^{y_i}}{\sum_{j}e^{y_j}}$$

```python
"""Softmax."""

scores = [3.0, 1.0, 0.2]

import numpy as np

def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    # TODO: Compute and return softmax(x)
    expo_x = np.exp(x)
    return expo_x/np.sum(expo_x,axis=0)


print(softmax(scores))

# Plot softmax curves
import matplotlib.pyplot as plt

x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```

เปลี่ยน x เป็น 10x

```python
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([10*x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```

จะเห็นว่า distribution ของ x ขยับห่างกันมากขึ้น ค่าที่ติดลบยิ่งมีสัดส่วนน้อย แล้วค่าที่บวกยิ่งใกล้ 1 หมด

เมื่อเปลี่ยน x เป็น x/10

```python
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x/10, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```

จะเห็นว่า distribution ของ x ขยับใกล้กันมากขึ้น จนเกือบจะกลายเป็น uniform distribution

### One-Hot Encoding 

หลังจากผ่าน softmax แล้วค่าที่ได้จะอยู่ระหว่าง 0-1 เช่น $S(y) =$ [0.7, 0.2, 0.1] ถ้าทำ one-hot encoding จะกลายเป็น $L =$ [1,0,0]

### Cross Entropy

one-hot encoding (ใช้ 0,1 label คำตอบ) จะให้ผลดี จนกว่าจะเจอกรณีที่ class เราเยอะๆ ทำให้ vector ใหญ่มากๆ ดังนั้นจะมี 0 อยู่เต็มไปหมด ซึ่งเราจะใช้วิธี embedding ในการแก้ปัญหานี้

the natural way to measure the distance between $S(y)$ กับ label($L$) is called the **Cross Entropy**

$$D(S,L) = -\sum_{i}L_i\text{log}(S_i)$$

***Note*** : *D(S,L) != D(L,D)*
